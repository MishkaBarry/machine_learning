```python
#Применим OheHot
def onehot(data):
    from numpy import array
    from sklearn.preprocessing import LabelEncoder
    from sklearn.preprocessing import OneHotEncoder

    values = array(data)
    label_encoder = LabelEncoder()
    integer_encoded = label_encoder.fit_transform(values)
    onehot_encoder = OneHotEncoder(sparse=False)
    integer_encoded = integer_encoded.reshape(len(integer_encoded), 1)
    onehot_encoded = onehot_encoder.fit_transform(integer_encoded)
    return onehot_encoded

encodedVehicleType = onehot(file2['vehicleType'])
dfVehicleType = pd.DataFrame(encodedVehicleType, columns=['coupe', 'suv', 'kleinwagen', 'limousine', 'cabrio','bus', 'veh_kombi', 'veh_andere' ])
dfVehicleType = dfVehicleType.astype(int)
dfVehicleType
```

```python
#Собираем итоговую матрицу признаков
train = np.hstack([dffuelType, X_brand, X_model, dfVehicleType, pd.DataFrame(file2['powerPS']),pd.DataFrame(file2['yearOfRegistration']),pd.DataFrame(file2["kilometer"]),pd.DataFrame(file2["notRepairedDamage"]),pd.DataFrame(file2["monthOfRegistration"]),pd.DataFrame(file2["gearbox"]),pd.DataFrame(file2["abtest"])])
train = pd.DataFrame(train)
train2 = train.set_axis(['benzin', 'diesel', 'lpg', 'andere', 'hybrid','cng','elektro','brand', 'model', 'coupe','suv','kleinwagen','limousine','cabrio','bus','veh_kombi','veh_andere', 'powerPS', 'yearOfRegistration','kilometer','notRepairedDamage','monthOfRegistration','gearbox','abtest'], axis=1, inplace=False)
train2

```

```python
#линейная регрессия
from sklearn.metrics import r2_score as r2, mean_squared_error as mse
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(train2, newprice, test_size=0.30, random_state=1)
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
y_pred = regressor.predict(X_test)

import numpy as np
from sklearn.metrics import mean_absolute_error
print(r2(y_test, y_pred))
print(mse(y_test, y_pred))
rmse_lin = mse(y_test, y_pred) *0.5
r2_lin = r2(y_test, y_pred)
```

```python
#регуляризация L2
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import Ridge

ridge = Ridge()
parameters = {'alpha': [5, 10, 12, 20]}
ridge_regressor = GridSearchCV(ridge, parameters, scoring='neg_mean_squared_error',  cv=5)
ridge_regressor.fit(train2, newprice)
y_pred = ridge_regressor.predict(X_test)
print(r2(y_test, y_pred))
print(mse(y_test, y_pred))
rmse_l2 = mse(y_test, y_pred) *0.5
r2_l2 = r2(y_test, y_pred)
```

```python
#регуляризация L1
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import Lasso

lasso = Lasso()
parameters = {'alpha': [0.01, 1, 5, 10, 20]}
lasso_regressor = GridSearchCV(lasso, parameters,scoring='neg_mean_squared_error',  cv=5)
lasso_regressor.fit(train2, newprice)
y_pred = lasso_regressor.predict(X_test)
print(r2(y_test, y_pred))
print(mse(y_test, y_pred))
rmse_l1 = mse(y_test, y_pred) *0.5
r2_l1 = r2(y_test, y_pred)
```

```python
#Случайный лес
from sklearn.ensemble import RandomForestRegressor
forestRegressor = RandomForestRegressor(n_estimators=10, random_state=0)
forestRegressor.fit(X_train, y_train)
y_pred = forestRegressor.predict(X_test)
print(r2(y_test, y_pred))
print(mse(y_test, y_pred))
rmse_rf = mse(y_test, y_pred) *0.5
r2_rf = r2(y_test, y_pred)
```

```python
#Градиентный бустинг
from sklearn import ensemble
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(train2, newprice, test_size=0.30, random_state=1)
model = ensemble.GradientBoostingRegressor()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
print(r2(y_test, y_pred))
print(mse(y_test, y_pred))
rmse_gb = mse(y_test, y_pred) *0.5
r2_gb = r2(y_test, y_pred)
```
